```python
# prompt: select features with correlation coefficient above a threshold , without considering the last column

import pandas as pd

# Assuming databf_features is a pandas DataFrame
threshold = 0.9  # Set the desired threshold

# Get the correlation matrix
correlation_matrix = databf_features.corr()

# Select features with high correlation
selected_features = []
for i in range(len(correlation_matrix.columns) - 1):
    for j in range(i + 1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            selected_features.append(correlation_matrix.columns[i])
            selected_features.append(correlation_matrix.columns[j])

# Remove duplicates
selected_features = list(set(selected_features))

# Print the selected features
print(f"Selected features: {selected_features}")

```

```python
from sklearn import model_selection

# Split data into train and test
S_Train_set,S_Test_set = model_selection.train_test_split(X_train_selected,test_size=0.3)
print("Shape of the newly selected dataset (X_train_selected) data: ", X_train_selected.shape)
print("Shape of the X_train_selected train data = ", S_Train_set.shape)
print("Shape of the X_train_selected test data = ", S_Test_set.shape)

#Resetting index for prediction time
#iris_train = iris_train.reset_index(drop=True)
#iris_test = iris_test.reset_index(drop=True)
```

```python
# prompt: now split the test and train sets in to features and labels

# Extract features and labels from train set
X_train_features = S_Train_set
y_train_labels = y_train[S_Train_set.index]

# Extract features and labels from test set
X_test_features = S_Test_set
y_test_labels = y_train[S_Test_set.index]
```

```python
rf_selected = RandomForestClassifier()
rf_selected.fit(X_train_features, y_train_labels)
```

```python
from sklearn.model_selection import cross_val_score , KFold

# Initialize the model
rf = RandomForestClassifier()

kf = KFold(n_splits=10, shuffle=True, random_state=1)
# Perform 10-fold cross-validation
cv_scores = cross_val_score(rf, X_train_features, y_train_labels, cv=kf, scoring='accuracy')

# Print cross-validation results
print("kfold:", kf)
print("Cross-validation scores:", cv_scores)
print("Mean cross-validation score:", cv_scores.mean())
```

```python
#Predictions on the Test data
predictions_rf_selected = rf_selected.predict(X_test_features)

#Putting the above predictions and probabilities together to create a readable df
#X_pred_label = pd.DataFrame(predictions_rf_selected.reshape(len(predictions_rf_selected),1))
#X_pred_label.rename(columns={0:'dt_prediction'}, inplace=True)
#X_pred_label.head(5)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Accuracy
accuracy = accuracy_score(y_test_labels, predictions_rf_selected)
print(f'Accuracy: {accuracy}')

# Precision
precision = precision_score(y_test_labels, predictions_rf_selected, average='weighted')  # Change 'weighted' to 'macro' or 'micro' based on your use case
print(f'Precision: {precision}')

# Recall
recall = recall_score(y_test_labels, predictions_rf_selected, average='weighted')
print(f'Recall: {recall}')

# F1-Score
f1 = f1_score(y_test_labels, predictions_rf_selected, average='weighted')
print(f'F1-Score: {f1}')

# Confusion Matrix
cm = confusion_matrix(y_test_labels, predictions_rf_selected)
print('Confusion Matrix:')
print(cm)

# Classification Report
report = classification_report(y_test_labels, predictions_rf_selected)
print('Classification Report:')
print(report)
```
